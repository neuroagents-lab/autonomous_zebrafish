```python
from stable_baselines3.common.vec_env import VecVideoRecorder, DummyVecEnv
from zfa_rl_agent.core.environments.zebrafish import swimmer_venv, swimmer
from zfa_rl_agent.core.environments.curiosity_venv_wrapper import curiosity_venv
from zfa_rl_agent.core.agent.PPO.rPPO import PPO
#from sb3
model_path = '/data/user_data/rdkeller/rl_training_logs/zfa_trained_defaults_swim0.0_idm1.0_2D-drift0.05-2025-03-03_18-18-46/checkpoints/zfa_trained_defaults_swim0.0_idm1.0_2D-drift0.05-2025-03-03_18-18-46_16694656_steps.zip'
video_folder = '/home/rdkeller/zebrafish_agent/zfa_rl_agent/experiments/continue_training/videos'
```

```python
eval_length = 1000
image_ego = []
image_top = []
force = 0.0
env = swimmer_venv(parallel=0, 
                   use_ray=0, 
                   n_envs=1, 
                   seed=1234, 
                   monitor_dir=None, 
                   n_frames=0, 
                   force_magnitude=force, 
                   training=True)

vis_env = swimmer_venv(parallel=0, 
                   use_ray=0, 
                   n_envs=1, 
                   seed=1234, 
                   monitor_dir=None, 
                   n_frames=0, 
                   force_magnitude=force, 
                   training=False)

model = PPO.load(model_path, env)

obs = model.env.reset()
vis_env.reset()
img_ego = model.env.render(mode="rgb_array")
img_top = vis_env.render(mode="rgb_array")
for _ in range(eval_length):
    image_top.append(img_top)
    image_ego.append(img_ego)
    action, _ = model.predict(obs)
    #print(np.linalg.norm(action))
    obs, _, _, _ = model.env.step(action)
    vis_env.step(action)
    img_ego = model.env.render(mode="rgb_array")
    img_top= vis_env.render(mode="rgb_array")

prefix = f"2D_force_{force}_trough_"
imageio.mimsave(video_folder + "/" + prefix + "swim_top.gif", [np.array(img) for i, img in enumerate(image_top) if i%2==0], fps=29)
imageio.mimsave(video_folder + "/" + prefix + "swim_ego.gif", [np.array(img) for i, img in enumerate(image_ego) if i%2==0], fps=29)
```
